# Generate Synthetic Data Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_data(n=50, noise=0.3):
    x = np.random.rand(n) * 10
    err = np.random.normal(size=n) * noise
    y = np.sin(x) + err
    return x, y
```

```python
def generate_synthetic_data(func, x_range, n_points, error):
    """
    Generate synthetic data based on a provided mathematical function.
    
    Parameters:
        - func (callable): The mathematical function to generate y-values.
        - x_range (tuple): The range of x-values (start, end).
        - n_points (int): Number of data points to generate.
        - error (float): Standard deviation of the Gaussian noise added to y-values.

    Returns:
        - tuple: Arrays of x-values and y-values.
    """

    x_values = np.linspace(x_range[0], x_range[1], n_points)
    y_values = func(x_values)

    # add Gaussian noise: mu=0, sigma=error
    y_values += np.random.normal(0, error, n_points)

    return x_values, y_values
```

```python
def visualize_data(x_values, y_values, ylabel):
    """
    Visualize the data using a scatter plot.
    
    Parameters:
        - x_values (array): Array of x-values.
        - y_values (array): Array of y-values.
    """
    plt.scatter(x_values, y_values)
    plt.xlabel('X-axis')
    plt.ylabel(ylabel)
    plt.title('Synthetic Data Visualization')
    plt.show()
```

```python
# define a linear function
def f(x):
    return 1.3 * x + 11.9

# define non-linear function
def non_linear(x):
    return np.sqrt(x) + 15

# defin transcedental
def sin(x):
    return np.sin(x)
```

```python
# generate linear synthetic data
x_vals, y_vals = generate_synthetic_data(f, (0, 100), 300, 8.4)

# Visualize the data
visualize_data(x_vals, y_vals, 'f(x)')
```

```python
# generate polynomial synthetic data
x_vals, y_vals = generate_synthetic_data(non_linear, (0, 40), 300, 5.2)

# Visualize the data
visualize_data(x_vals, y_vals, 'np.sqrt(x) + 15')
```

```python
# generate polynomial synthetic data
x_vals, y_vals = generate_synthetic_data(sin, (0, 50), 100, 1.0)

# Visualize the data
visualize_data(x_vals, y_vals, 'sin(x)')
```
